```python
# Menu à afficher
affichage = """
Choisissez une option:
    1: Ajouter un élément
    2: Enlever un élément
    3: Afficher la liste
    4: Vider la liste
    5: Sauvegarder la liste
    0: Quitter
"""

# Fonctions -----------------------------------------------

def ajouter_element(liste):
    element = input("Entrez l'élément à ajouter : ")
    liste.append(element)
    print(f"L'élément '{element}' a été ajouté.")

def enlever_element(liste):
    element = input("Entrez l'élément à enlever : ")
    if element in liste:
        liste.remove(element)
        print(f"L'élément '{element}' a été retiré.")
    else:
        print("Élément non trouvé dans la liste.")

def afficher_liste(liste):
    if liste:
        print("Liste actuelle :", liste)
    else:
        print("La liste est vide.")

def vider_liste(liste):
    liste.clear()
    print("La liste a été vidée.")

def sauvegarder_liste(liste):
    import json
    with open("ma_liste.json", "w") as f:
        json.dump(liste, f, indent=4)
    print("Liste sauvegardée dans 'ma_liste.json'.")

# Programme principal --------------------------------------

ma_liste = []

while True:
    print(affichage)
    choix = input("Votre choix : ")

    if choix == "1":
        ajouter_element(ma_liste)
    elif choix == "2":
        enlever_element(ma_liste)
    elif choix == "3":
        afficher_liste(ma_liste)
    elif choix == "4":
        vider_liste(ma_liste)
    elif choix == "5":
        sauvegarder_liste(ma_liste)
    elif choix == "0":
        print("Au revoir !")
        break
    else:
        print("Option invalide, veuillez réessayer.")


print(ma_liste)
```
